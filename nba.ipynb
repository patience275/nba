## recomendation system

```python
player_info=df[['Player','Tm','MP','gmsc/36']]
```

```python
perfomance=df[['FGA', 'FG%', '3P', '3PA', '3P%', 'FT',
       'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF']]
```

```python
perfomance.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
perfomance=pd.DataFrame(scaler.fit_transform(perfomance),columns=perfomance.columns)
perfomance=perfomance.round(2)
perfomance.head()
```

```python
scout_needs={
    '3P':5,
    'AST':3
}
```

```python
import numpy as np
input_vector=[]
for feature in perfomance:
  if feature in scout_needs:
    value=scout_needs[feature]
    row=[]
    for f in perfomance:
      if f==feature:
        row.append(value)
      else:
        row.append(0)
    scaled_value=scaler.transform([row])[0][perfomance.columns.get_loc(feature)]
    input_vector.append(scaled_value)
  else:
    input_vector.append(0.0)

input_vector = np.array(input_vector).reshape(1, -1)
```

```python
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np

# Calculate Euclidean distances
distances = euclidean_distances(perfomance, input_vector)

# Add distances to the recommendation dataframe
recomendation_euclidean = player_info.copy()
recomendation_euclidean['distances'] = distances

# Sort by distances in ascending order (smaller distance means more similar)
recomendation_euclidean = recomendation_euclidean.sort_values(by='distances', ascending=True)

# Display the top 10 recommendations based on Euclidean distance
display(recomendation_euclidean.head(10))
```

```python
# Merge the recommendation dataframe with the original dataframe to get the 'undervalued' status
undervalued_recs = pd.merge(recomendation_euclidean, df[['Player', 'undervalued']], on='Player')

# Filter for undervalued players
undervalued_recs = undervalued_recs[undervalued_recs['undervalued'] == True]

# Display the top undervalued recommendations
display(undervalued_recs.head(10))
```

```python
undervalued_recs_unique = undervalued_recs.drop_duplicates(subset=['Player'])
display(undervalued_recs_unique.head(10))
```
