```python
x=old_df[[ 'Age',  'G', 'GS', 'MP', 'PER',
       'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%',
       'TOV%', 'USG%', 'OWS', 'DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM',
       'VORP']]
x.head()
```

```python
player_info=old_df[['Player', 'Team', 'Pos','Awards']]
player_info.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x.head()
```

```python
scout_input={
    'Age':24,
    'DBPM':2.5,
    'stl%':2.0,
    'BLK%':2.0,
    'AST%':20.0
}
```

```python
import numpy as np

all_features = x.columns
# Create a dictionary with all features and their corresponding values from scout_input
input_data = {feature: scout_input.get(feature, 0.0) for feature in all_features}

# Convert the dictionary to a DataFrame with a single row
input_df = pd.DataFrame([input_data])

# Scale the input DataFrame using the fitted scaler
input_vector = scaler.transform(input_df)

print(input_vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities=cosine_similarity(input_vector,x)[0]
top_indices=similarities.argsort()[::-1][10]
recomendation=player_info.iloc[top_indices].copy()
recomendation['similarities']=similarities[top_indices]
recomendation
```

```python
from flask import Flask,request,render_template_string
from flask_ngrok import run_with_ngrok
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

app=Flask(__name__)
HTML_TEMPLATE='''
<!doctype html>
<title>hoopscout:intelligent basketball scout recomender</title>
<h2>player recommendation</h2>
<form method='post'>
  Age:<input name='Age' type='number' step='any'><br>
  DBPM:<input name='DBPM' type='number' step='any'><br>
  STL%:<input name ='STL%' type='number' step='any'><br>
  BLK%:<input name ='BLK%' type='number' step='any'><br>
  AST%:<Iinput name='AST%' type='number' step='any'><br>
  PER%:<Iinput name='PER' type='number' step='any'><br>
  TS%:<Iinput name='TS%' type='number' step='any'><br>
  3PAr:<Iinput name='3PAr' type='number' step='any'><br>
  FTr:<Iinput name='FTr' type='number' step='any'><br>
  ORB%:<Iinput name='ORB%' type='number' step='any'><br>
  DRB%:<Iinput name='DRB%' type='number' step='any'><br>
  TRB%:<Iinput name='TRB%' type='number' step='any'><br>
  TOV%:<Iinput name='TOV%' type='number' step='any'><br>
  USG%:<Iinput name='USG%' type='number' step='any'><br>
  OWS:<Iinput name='OWS' type='number' step='any'><br>
  DWS:<Iinput name='DWS' type='number' step='any'><br>
  WS:<Iinput name='WS' type='number' step='any'><br>
  WS/48:<Iinput name='WS/48' type='number' step='any'><br>
  OBPM:<Iinput name='OBPM' type='number' step='any'><br>
  VORP:<Iinput name='VORP' type='number' step='any'><br>
  <input type ='submit' value='recommend'>
</form>

{% if recommendations %}
  <h3>Top Recommendations</h3>
  <table border="1">
    <tr><th>Player</th><th>Team</th><th>Pos</th><th>Similarity</th></tr>
    {% for r in recommendations %}
    <tr>
      <td>{{ r['Player'] }}</td>
      <td>{{ r['Team'] }}</td>
      <td>{{ r['Pos'] }}</td>
      <td>{{ '{:.2f}'.format(r['Similarity']) }}</td>
    </tr>
    {% endfor %}
  </table>
{% endif %}
'''
@app.route('/',methods=['GET','POST'])
def recommend():
  recomendation=None
  if request.method=='POST':
    scout_input={k: float(request.form[k]) for k in['Age', 'G', 'GS', 'MP', 'PER', 'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%',
       'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%', 'OWS', 'DWS', 'WS',
       'WS/48', 'OBPM', 'DBPM', 'BPM', 'VORP']}

    all_features=list(x.columns)
    input_vec=[]
    for feature in all_features:
      if feature in scout_input:
        temp=[scout_input[feature] if f==feature else 0 for f in all_features]
        scaled_val=scaler.transform([temp])[0][all_features.index(feature)]
        input_vec.append(scaled_val)
      else:
        input_vec.append(0.0)
    input_vec=np.array(input_vec).reshape(1,-1)

    sims=cosine_similarity(input_vec,x)[0]
    top_idx=sims.argsort()[::-1][:10]
    rec=player_info.iloc[top_idx].copy()
    rec['similarity']=sims[top_idx]
    recomendations=rec.to_dict('records')
  return render_template_string(HTML_TEMPLATE,recomendations=recomendations)

if __name__=='__main__':
  app.run(debug=True)
```
