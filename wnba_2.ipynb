```python
metrics=['Age', 'PER',
       'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%',
       'TOV%', 'USG%', 'OWS', 'DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM']

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled_metrics=pd.DataFrame(scaler.fit_transform(df[metrics]),columns=metrics)
```

```python
scaled_metrics.head()
```

```python
players_info=['Player','Team', 'Pos', 'VORP', 'Awards', 'undervalued']
```

```python
players_info_df = df[['Player', 'Team', 'Pos', 'VORP', 'Awards', 'undervalued']]
```

```python
scaled_metrics.columns
```

```python
df.describe()[metrics]
```

```python
scout_needs={
    'AST%':15.0,
    'DRB%':12.0
}
needs_importance={
    'AST%': 1.0,
    'DRB%':0.7
}
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity=cosine_similarity
mean_values=df[metrics].mean()
filled_scout={}

for feat in metrics:
    if feat in scout_needs:  # If scout gave a value for this stat
        filled_scout[feat] = scout_needs[feat]
    else:  # If scout didn't mention it, use the mean
        filled_scout[feat] = mean_values[feat]

scout_vector = pd.DataFrame([filled_scout])[metrics]
scout_vector_scaled=scaler.transform(scout_vector)

weights_arr = np.array([needs_importance.get(f, 1.0) for f in metrics])
all_weighted = scaled_metrics * weights_arr
scout_weighted = scout_vector_scaled[0] * weights_arr
scout_weighted = scout_weighted.reshape(1, -1)


sim_scores = cosine_similarity(scout_weighted, all_weighted)

# Flatten sim_scores to 1D array
df['Similarity'] = sim_scores.flatten()
results = df.sort_values('Similarity', ascending=False)

print(results[players_info])

```
